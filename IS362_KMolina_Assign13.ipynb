### Introduction
In this assignment, we aim to preprocess a mock Mushroom dataset inspired by the UCI Mushroom Dataset. The dataset contains information about mushrooms, including whether they are edible or poisonous, their odor, and other features. This preprocessing involves:
1. Creating and understanding the dataset.
2. Subsetting and renaming columns with meaningful names.
3. Converting categorical data into numeric values for compatibility with machine learning models.
4. Performing exploratory data analysis (EDA) to visualize the relationships between features.
5. Drawing preliminary conclusions about feature importance.

### Step 1: Importing Required Libraries
We start by importing the necessary Python libraries:
- **`pandas`**: For creating and manipulating the dataset.
- **`matplotlib.pyplot`** and **`seaborn`**: For visualizing the dataset and performing exploratory data analysis (EDA).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensuring plots are displayed inline in the notebook
%matplotlib inline


````python
import matplotlib.pyplot as plt
````

````python
# Importing Pandas
import pandas as pd

# Simulating a dataset
mock_data = {
    'class': ['e', 'p', 'e', 'p', 'e', 'p', 'p', 'e'],  # e=edible, p=poisonous
    'odor': ['a', 'l', 'a', 'n', 'c', 'y', 'f', 's'],   # odors: a=almond, l=licorice, etc.
    'cap-color': ['n', 'y', 'w', 'n', 'g', 'n', 'y', 'w']  # cap colors: n=brown, y=yellow, etc.
}

# Creating a DataFrame
data = pd.DataFrame(mock_data)

# Display the mock dataset
data

````

````python
# Subset the selected columns
subset = data[['class', 'odor', 'cap-color']]

# Rename columns
subset.columns = ['is_edible', 'mushroom_odor', 'cap_color']

# Display the first few rows of the renamed subset
subset
````

````python
# Mapping values for the 'is_edible' column
subset['is_edible'] = subset['is_edible'].map({'e': 1, 'p': 0})

# Encoding 'mushroom_odor' and 'cap_color' as numeric
subset['mushroom_odor'] = subset['mushroom_odor'].astype('category').cat.codes
subset['cap_color'] = subset['cap_color'].astype('category').cat.codes

# Display the first few rows of the transformed dataset
subset
````

````python
# Importing required libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of 'is_edible'
plt.figure(figsize=(6, 4))
sns.countplot(x='is_edible', data=subset)
plt.title('Distribution of Edible vs. Poisonous Mushrooms')
plt.xlabel('Is Edible (1=Edible, 0=Poisonous)')
plt.ylabel('Count')
plt.show()

# Plot distribution of 'mushroom_odor'
plt.figure(figsize=(6, 4))
sns.countplot(x='mushroom_odor', data=subset)
plt.title('Distribution of Mushroom Odor')
plt.xlabel('Odor (Encoded)')
plt.ylabel('Count')
plt.show()

# Plot distribution of 'cap_color'
plt.figure(figsize=(6, 4))
sns.countplot(x='cap_color', data=subset)
plt.title('Distribution of Cap Color')
plt.xlabel('Cap Color (Encoded)')
plt.ylabel('Count')
plt.show()

````

#### Observations:
1. Mushroom odor shows a strong correlation with edibility, as certain odors are more likely to indicate poisonous mushrooms.
2. Cap color shows some variation between edible and poisonous mushrooms, but it appears to be less predictive than odor.

#### Recommendations:
- Use mushroom odor as a critical feature for any predictive model.
- Cap color may serve as a supporting feature but has less predictive power than odor.


````python
# Save the processed dataset to a CSV file
processed_file_path = "processed_mushroom_data.csv"
subset.to_csv(processed_file_path, index=False)

print(f"Processed data saved to {processed_file_path}")
````
